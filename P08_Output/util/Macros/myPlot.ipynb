```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("electron1GeV.csv", comment = '#', header=None)
#data.head()

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(2,2,1)


ax1.grid(True)
ax1.set_xlabel("E deposit(MeV)")
ax1.set_ylabel("Step length (mm)")
ax1.set_title("proton into BGO")
#ax1.set_ylim(8000, 12000.)
#ax1.set_xlim(800, 1000)
ax1.scatter(data[0], data[1], s=1, marker= '.')

ax2 = fig.add_subplot(2,2,2)
ax2.set_ylabel("Step length (mm)")
ax2.set_xscale('log')
ax2.hist(data[1], bins=100,orientation='horizontal')  #, range=[8000., 12000.])

ax3 = fig.add_subplot(2,2,3)
ax3.set_yscale('log')
ax3.set_xlabel("Energy Deposit (MeV)")
ax3.hist(data[0], bins=100) #, range=[800,  1000])
```
